```python
import matplotlib.pyplot as plt
import soundfile as sf
import numpy as np

def wav_to_spectrogram(audio_path, save_path, spectrogram_dimensions=(64, 64), noverlap=16, cmap='gray_r'):
    try:
        samples, sample_rate = sf.read(audio_path)
    except RuntimeError as e:
        print(f"Error reading {audio_path}: {e}")
        return

    if samples.ndim > 1:
        samples = samples.mean(axis=1)  # convert to mono

    fig = plt.figure()
    fig.set_size_inches((spectrogram_dimensions[0]/fig.get_dpi(), spectrogram_dimensions[1]/fig.get_dpi()))
    ax = plt.Axes(fig, [0., 0., 1., 1.])
    ax.set_axis_off()
    fig.add_axes(ax)
    Pxx, freqs, bins, im = ax.specgram(samples, NFFT=256, Fs=sample_rate, noverlap=noverlap, cmap=cmap)
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())
    fig.savefig(save_path, bbox_inches="tight", pad_inches=0)
    plt.close(fig)
```

```python
audio_path=[]
def dir_to_spectrogram(audio_dir, spectrogram_dir, spectrogram_dimensions=(64, 64), noverlap=16, cmap='gray_r'):
  file_names = [f for f in listdir(audio_dir) if isfile(join(audio_dir, f)) and '.wav' in f]
  for file_name in file_names:
    print(file_name)
    audio_path = audio_dir + file_name
    spectogram_path = spectrogram_dir + file_name.replace('.wav', '.png')
    wav_to_spectrogram(audio_path, spectogram_path, spectrogram_dimensions=spectrogram_dimensions, noverlap=noverlap, cmap=cmap)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
#from tensorflow.keras.preprocessing import image
#from keras.utils import to_categorical
from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import to_categorical
#from keras.utils.np_utils import to_categorical
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Define paths
dataset_folder = "/content/drive/MyDrive/CNN/Spectrogram/"
train_folder = "/content/drive/MyDrive/CNN/Train/"
test_folder = "/content/drive/MyDrive/CNN/Test/"

# Create training and testing folders if they don't exist
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# List all files in the dataset folder
files = os.listdir(dataset_folder)

# Split the dataset into training and testing sets
train_files, test_files = train_test_split(files, test_size=0.2, random_state=42)

# Move training files to the training folder
for file in train_files:
    shutil.move(os.path.join(dataset_folder, file), os.path.join(train_folder, file))

# Move testing files to the testing folder
for file in test_files:
    shutil.move(os.path.join(dataset_folder, file), os.path.join(test_folder, file))

```

```python
trainset= []
train_directory = "/content/drive/MyDrive/CNN/Train/"
for file in os.listdir(train_directory):
  labels = file.split('_')[0]
  sample_number = file.split('_')[2]
  img = image.load_img(train_directory+file)
  trainset.append([image.img_to_array(img), labels])
```

```python
testset= []
test_directory = "/content/drive/MyDrive/CNN/Test/"
for file in os.listdir(test_directory):
  labels = file.split('_')[0]
  sample_number = file.split('_')[2]
  img = image.load_img(test_directory+file)
  testset.append([image.img_to_array(img), labels])
```

```python
#
# Get only images in the train list not the Labels
X_train = [item[0] for item in trainset]
# Get only Labels in the train list not the images
y_train = [item[1] for item in trainset]
# Get only images in the test list not the Labels
X_test = [item[0] for item in testset]
# Get only Labels in the test list not the images
y_test = [item[1] for item in testset]
```

```python
# Convert list to numpy array in order to define input shape
X_train = np.asanyarray(X_train)
y_train = np.asanyarray(y_train)
X_test = np.asanyarray(X_test)
y_test = np.asanyarray(y_test)
print("X_train",X_train.shape)
print("y_train",y_train.shape)
print("X_test",X_test.shape)
print("y_test",y_test.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
# Assuming y_train is a list of string labels
label_encoder = LabelEncoder()
#y_train_encoded = label_encoder.fit_transform(y_train)

# Now y_train_encoded contains integer labels

y_test_encoded = label_encoder.fit_transform(y_test)
y_train_encoded = label_encoder.fit_transform(y_train)

# Reshape y_train_encoded to 1D array
#y_train_encoded = np.argmax(y_train_encoded, axis=1)

# Now y_train_encoded contains integer labels

y_train_one_hot = to_categorical(y_train_encoded)
y_test_one_hot = to_categorical(y_test_encoded)






```

```python
import numpy as np

# Assuming X_train and X_test are numpy arrays representing images and y_train, y_test are the corresponding labels

# Convert to float
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# Normalize the images
X_train /= 255
X_test /= 255
```

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from keras import optimizers
from keras.layers import Dense, Activation, Flatten, Conv2D, MaxPooling2D, Dropout, BatchNormalization
from keras import models
from keras.optimizers import SGD
from keras import regularizers
```

```python
data_shape = (X_train.shape[1], X_train.shape[2], X_train.shape[3])


def basic_cnn():
  model = Sequential()
  model.add(Conv2D(32, kernel_size=(2, 2), activation='relu', input_shape=data_shape))
  model.add(BatchNormalization())
  model.add(Conv2D(48, kernel_size=(2, 2), activation='relu'))
  model.add(BatchNormalization())
  model.add(Conv2D(120, kernel_size=(2, 2), activation='relu'))
  model.add(BatchNormalization())
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(0.25))
  model.add(Flatten())
  model.add(Dense(128, activation='relu'))
  model.add(BatchNormalization())
  model.add(Dropout(0.25))
  model.add(Dense(64, activation='relu'))
  model.add(BatchNormalization())
  model.add(Dropout(0.4))
  model.add(Dense(10, activation='softmax'))
  optimizer = SGD()
  model.compile(loss = 'categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
  return model





```

```python
model0 = basic_cnn()
model0.summary()
```

```python

# Check data shapes
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train_one_hot.shape)
model0.fit(X_train, y_train_one_hot, batch_size = 50, validation_split=0.2, epochs = 100, verbose = 1)










```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import tensorflow as tf
from matplotlib.font_manager import FontProperties
from tensorflow import keras

# Font properties for labels
font_prop = FontProperties(fname="/content/drive/MyDrive/CNN/Mangal Regular.ttf", size=12)

# Function to calculate TP, TN, FP, FN
def calculate_confusion_metrics(confusion_mat):
    tp = np.diag(confusion_mat)
    fp = np.sum(confusion_mat, axis=0) - tp
    fn = np.sum(confusion_mat, axis=1) - tp
    tn = np.sum(confusion_mat) - (tp + fp + fn)
    return tp, tn, fp, fn

# Function to calculate precision, recall, accuracy, and F1-score
def calculate_metrics(tp, tn, fp, fn):
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    f1_score = 2 * (precision * recall) / (precision + recall)
    return precision, recall, accuracy, f1_score

# Assuming y_pred and y_test are already defined
# y_pred = model.predict(X_test)
# y_pred_classes = np.argmax(y_pred, axis=1)
# y_true_classes = np.argmax(y_test, axis=1)

# Calculate the confusion matrix
confusion_mat = confusion_matrix(y_true_classes, y_pred_classes)

# Calculate TP, TN, FP, FN for each class
tp, tn, fp, fn = calculate_confusion_metrics(confusion_mat)

# Calculate precision, recall, accuracy, and F1-score for each class
precision, recall, accuracy, f1_score = calculate_metrics(tp, tn, fp, fn)

# Display the results
class_labels = ["शून्य", "एक", "दो", "तीन", "चार", "पाँच", "छह", "सात", "आठ", "नौ"]

for i, label in enumerate(class_labels):
    print(f"Class '{label}':")
    print(f"  True Positives: {tp[i]}")
    print(f"  True Negatives: {tn[i]}")
    print(f"  False Positives: {fp[i]}")
    print(f"  False Negatives: {fn[i]}")
    print(f"  Precision: {precision[i]:.4f}")
    print(f"  Recall: {recall[i]:.4f}")
    print(f"  Accuracy: {accuracy[i]:.4f}")
    print(f"  F1-Score: {f1_score[i]:.4f}")
    print()

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_mat, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.xticks(fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
